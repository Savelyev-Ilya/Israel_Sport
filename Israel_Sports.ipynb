## Data cleaning

To begin, I will look for the unique values in all of the required columns to see how much I should translate. I am lazy enough to just make a new dataset with the columns I need and output all the values in a time with the "for" cycle.

```python
df_unique = df.drop(["Id", "X", "Y"], axis = 1)
for i in df_unique.columns:
    print("Unique values of the column ", i,":")
    print(df_unique[i].unique())
    print()
```

So, what I have here. The list of the cities is extremely massive, so I hope you forgive me for not translating it :)

Anyway, I believe that those who will be seriously interested in the results should know Hebrew. Moreover, I am going to translate those cities which will be significant for the results and visualisations.

All the other columns besides the "City" and "Type" can be translated easily, which I am going to do right now.

Also I can see some missing values in the "Status", "Compatitions" and "Year of building" columns. I am going to deal with them a bit later.

```python
df1 = df
df1['Parking'] = df1['Parking'].replace(['לא', 'כן'],['No', 'Yes'])
df1['Accessible for disabled'] = df1['Accessible for disabled'].replace(['לא', 'כן'],['No', 'Yes'])
df1['Lighting'] = df1['Lighting'].replace(['לא', 'כן'],['No', 'Yes'])
df1['Fencing'] = df1['Fencing'].replace(['לא', 'כן'],['No', 'Yes'])
df1['Status'] = df1['Status'].replace(['תקין ופעיל', 'פעיל וזקוק לשיפוץ', 'לא פעיל', 'לא תקין ולא פעיל','סגור לשימוש'],
                                      ['Normal and active', 'Active and in need of renovation', 'Inactive', 'Abnormal and inactive','Closed to Use'])
df1['Competitions'] = df1['Competitions'].replace(['ליגה מקומית ואיזורית', 'מתקיימות במתקן תחרויות רשמיות','אינו משמש לתחרויות רשמיות', 'לא רלוונטי', 'ליגת על','תקני במידות אך ללא שירותים, מקלחות וחדר שופטים','משחקים בינלאומיים'],
                                                                    ['Local and regional', 'National', 'Not used for official competitions', 'Irrelevant', 'Super League', 'Standardised but without toilets, showers and a ref room', 'International Games'])
df1.head()
```

Now the "Type" column. It is related to the venues that were built for different types of sports, so I am going to translate them according to the keywords.

```python
df1 = df1.replace('^.*כדורגל.*$', "Football field", regex = True)
df1 = df1.replace('^.*כדורסל.*$', "Basketball court", regex = True)
df1 = df1.replace('^.*אולם ספורט.*$', "Fitness", regex = True)
df1 = df1.replace('^.*לכושר גופני.*$', "Fitness", regex = True)
df1 = df1.replace('^.*בריכת.*$', "Swimming pool", regex = True)
df1 = df1.replace('^.*טניס.*$', "Tennis court", regex = True)
df1 = df1.replace('^.*אחר.*$', "Other", regex = True)
df1 = df1.replace('^.*מגרש חול קבוע.*$', "Sand court", regex = True)
df1 = df1.replace('^.*אצטדיון.*$', "Stadium track", regex = True)
df1 = df1.replace('^.*מוטורי.*$', "Racing track", regex = True)
df1 = df1.replace('^.*אקסטרים.*$', "Extreme", regex = True)
df1 = df1.replace('^.*שייט מרכז ימי.*$', "Sea centre cruise", regex = True)
df1 = df1.replace('^.*מגרש ספורט משולב.*$', "Combined sports pitch", regex = True)
df1 = df1.replace('^.*מיני פיץ.*$', "Mini pitch", regex = True)
df1 = df1.replace('^.*דשא סינטטי.*$', "Synthetic grass court", regex = True)
df1 = df1.replace('^.*פטאנק.*$', "Petanque field", regex = True)
df1.head()
```

Great. Now I am going to solve the problem of the missing values. First of all, I would like to see how many missing values are there in the each column.

```python
df1.isnull().sum()
```

Here I can see that there is only one missing value in the "Status" column and I am not going to drop it, because it can be valuable in terms of the other characteristics.

However, not everything is so fine with the "Competitions" and "Year of building". The last one looks like the most controversial, as it is almost one-fourth of the dataset and dropping it may strongly affect the results. On the other hand, it's difficult to replace "Nan" cells with a particular value as it can also bias the results significantly. I think that it is better to keep it as it is in this case and just keep in mind that the question №7 may not be answered properly with the given dataset.

What about the "Competitions". I need to see which types of sports are related to these rows to be able to make a decision.

```python
df1[df1["Competitions"].isnull() == 1]["Type"].value_counts()
```

Now I will think in the following way:

Some of the categories are rarely used for competitions. The others are used oftenly, however, as they were not assigned to those ones, that are suitable for competitons, it is possible to assume that they are actually not. 

Finally, I decided to assign the value "Not used for official competitions" for the empty cells.

```python
df1["Competitions"] = df1["Competitions"].replace(np.nan, "Not used for official competitions", regex = True)
df1.isnull().sum()
```

Cool. Now the problems with missing values are more or less solved and I can move on to checking if there are proper datatypes for all the columns.

```python
df1.dtypes
```

As it can be seen. All the columns except the last one have proper types of data, so I need to alter only the "Year of building column". I want to get rid of ".0" which does not make any sence and I need to transform it from float to integer. However, since I have "some" missing values I firstly need to decide what to do with them. For now I will just assign "0" to all of them for easier filtering in the future.

```python
df1["Year of building"] = df1["Year of building"].fillna(0).astype("int")
df1.head()
```

Finally, my dataset is cleaned and prepared for the future analysis. Now I will just download it and I can begin answering the questions I asked at the beginning!

```python
df1.to_csv("Israel_sports.csv")
```

### 2. How many sports venues are there in Israel by types of sports?

```python
df_type = df1.groupby(["Type"])["Id"].count().reset_index()
df_type.sort_values(by = "Id", ascending = False, ignore_index = True)
```

As we can see here the greatest number of facilities belongs to fitness, which really makes sense because it is not so difficult and costly to build the venue of this purpose and many people use it. What seems really interesting is the fact that the number of basketball courts <b>more that two times outweighs</b> the number of football fields. However, this also makes sense as the courts for basketball are smaller that football fields, are easier to build and are usually located indoors, thus it is easier to look after them. Moreover, I have never thought that Petanque is relatively popular in Israel with <b>65</b> specialised fields for them.

### 3. Which cities have the biggest number of sports facilities?

```python
df_cities = df1.groupby(["City"])['Id'].count()
df_cities = pd.DataFrame(df_cities)
df_cities = df_cities.sort_values(by = "Id", ascending = False).reset_index()
df_cities.head(10)
```

As we can see the leaders are Jerusalem and Tel Aviv, which is natural, because these two cities are the biggest ones. However, besides the absolute values I am also interested in comparing the ratio of the cities' population to the number of facilities, so that the lowest values will indicate the cities that provide the highest sports accessibility for their people.

Of course, I will translate the names of the cities into English and also will join another dataset with the population data for these cities from the approximately same period. I do not promise the figures will be 100% precise, but I am now more interested in the overall picture.

```python
df_cities_top10 = df_cities.head(10)
df_pop = pd.read_csv("/Users/Savelyev_Ilya/Desktop/populations.csv")
df_join = pd.merge(df_cities_top10, df_pop, how = 'left', on = 'City')
df_join["City"] = df_join["City"].replace(['ירושלים', 'תל אביב-יפו', 'באר שבע', 'חיפה', 'נתניה', 'ראשון לציון', 'אילת', 'רחובות', 'אשקלון', 'אשדוד'],['Jerusalem', 'Tel Aviv-Yaffo', 'Beer Sheva', 'Haifa', 'Netanya', 'Rishon LeZion', 'Eilat', 'Rehovot', 'Ashkelon', 'Ashdod'])
df_join["People per venue"] = (df_join["Population"]/df_join["Id"]).astype("int")
df_join.sort_values(by = "People per venue", ignore_index = True)
```

So, it is clearly seen how dramatically the ranking changed. Eilat is the absolute leader and it really makes sence since it is a small city with a lot of hotels and tourists and most of them provide sports facilities usually. Contrariwise, even though Jerusalem has way more sports facilities that anyone else, its ratio of around 1311 people per venue is not so optimistic.

Let's visualize both findings!

```python
X = df_join["City"]
Y = df_join["Id"]
plt.figure(figsize=(15, 5))
plt.grid(color='grey', linestyle='-', alpha = 0.5, zorder = 0)
plt.bar(X, Y, zorder = 3)
plt.ylabel("Number of venues")
plt.suptitle('Top 10 cities by the total number of sports venues', size = 30)
plt.show()

X = df_join["City"]
Y = df_join["People per venue"]
plt.figure(figsize=(15, 5))
plt.grid(color='grey', linestyle='-', alpha = 0.5, zorder = 0)
plt.bar(X, Y, color = 'orange', zorder = 3)
plt.ylabel("People per venue")
plt.suptitle('Sports venues accessibility of the top 10 cities', size = 30)
plt.show()
```

### 5. How many sports venues currently need renovation?

Let's group all the venues by their status and look at the figures.

```python
df_status = df1.groupby(["Status"])['Id'].count().reset_index()
df_status
```

Looks like the most of the facilities are active and do not need a repair, however there are more than <b>1000</b> of active constructions that require renovation, which is a serious number!

I would also like to look at the figures among the venues that are suitable for the high-level competitions and then compare two charts together. So, for now I will just save the visualisation of the previous findings as a function and will find the values for the "International Games", "Super League" and "National" competitions facilities.

```python
def status():
    df_status["Id"].plot(kind='pie',
                   figsize=(10, 6),
                   autopct='%1.1f%%', 
                   startangle=90,
                   shadow = True,
                   ylabel = '',
                   labels=None,         
                   pctdistance=1.25,
                   colors = ['gold', 'yellowgreen', 'black', 'lightskyblue', 'lightcoral'],
                   explode=[0, 0, 0.1, 0, 0]
                  )
    plt.title('Sports venues distribution by status', y=1.12, size = 12)
    plt.axis('equal') 
```

```python
df_status_high = df1[df1.Competitions.isin(["International Games", "Super League", "National"])]
df_status_high = df_status_high.groupby(["Status"])['Id'].count().reset_index()
df_status_high
```

Here the situation is way more optimistic. While in general the ratio of proper venues to those that require renovation is about 7, in case of the high-level competitions it is more than 10.

Now it is time for visualization.

```python
def status_high():
    df_status_high["Id"].plot(kind='pie',
                     figsize=(10, 6),
                     autopct='%1.1f%%', 
                     startangle=90,
                     shadow = True,
                     ylabel = '',
                     labels=None,         
                     pctdistance=1.25,
                     colors = ['gold', 'yellowgreen', 'black', 'lightskyblue', 'lightcoral'],
                     explode=[0.1, 0, 0.1, 0, 0] 
                    )
    plt.title('High-level sports venues distribution by status', y=1.12, size = 12)
    plt.axis('equal')
    plt.legend(labels=df_status.Status, loc = 'lower left', fontsize=7) 
```

```python
plt.subplot(1,2,1)
status()
plt.subplot(1,2,2)
status_high()
plt.show()
```

What also looks great is the fact that there are less than 4% of the facilities that are not currently active. Looks like the sports constructions are used quite effectively in Israel!

### 6. What is the level of sport accessibility for disabled people?

It is clear that more than a half of all sports venues are accessible for disabled people, which I believe is a great proportion. If we also keep in mind that some types of sports facilities are impossible to make disabled-friendly because of the particular type of sport, we can conclude that the absolute majority of relevant sports objects are adapted to disabled.

```python
df_disabled = df1.groupby(["Accessible for disabled"])['Id'].count()
df_disabled = pd.DataFrame(df_disabled).reset_index()
df_disabled
```

```python
X = df_disabled["Accessible for disabled"]
Y = df_disabled['Id']
plt.figure(figsize=(15, 8))
plt.grid(color='grey', linestyle='-', alpha = 0.5, zorder = 0)
plt.bar(X, Y, color = ('red', 'green'), zorder = 3)
plt.ylabel("Number of venues")
plt.xlabel("Accessible for disabled")
plt.suptitle('Venues distribution by disable accessibility', size = 30)
plt.show()
```

### 7. How many sports facilities were built in Israel year-by-year?

Firstly, I will group the dataset by year as count the number of objects for each one.

```python
df_year = df1.groupby(["Year of building"])["Id"].count().reset_index()
df_year
```

And this is the moment, where the problem of the missing values occurs. Almost one-quarter of the dataset does not have a year and I have to decide, what to do with them.

I already thought about the problems at the data cleaning process and I think that the missing values should mostly belong to the older periods when there could be problems with handling the massive data. So, for now I will just exclude this values from the analysis, but keep in ming that in reality the graph will probably be "higher" on its right side.

```python
df_year_cleaned = df_year[df_year["Year of building"]!= 0]
X = df_year_cleaned['Year of building']
Y = df_year_cleaned['Id']
plt.figure(figsize=(15, 8))
plt.grid(color='grey', linestyle='-', alpha = 0.5)
plt.plot(X, Y)
plt.ylabel("Number of venues")
plt.xlabel("Year")
plt.suptitle('Number of venues built by year', size = 30)
plt.show()
```

Well, we have unusual figures! I didn't expect to be honest, that there will be so many peaks on the graph whith an enormous rise in 1970. It is interesting that there was a sharp increase in the number of sports facilities every 10th year from 1950 to 2000. I wonder how the graph could change if I didn't have so many missing values. Hope that one day the missing values will be corrected by the government.
